```python
import pandas as pd # biblioteca para manipulação e análise de dados
import numpy as np # biblioteca para manipulação matemática
```

```python
# Ferramentas de pré-processamento do scikit-learn
from sklearn.preprocessing import LabelEncoder, StandardScaler
# LabelEncoder -> Converte categorias para números
# StandardScaler -> Normaliza os dados (mesma escala)
```

```python
# Montar o Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/train.csv')
df.head()
```

```python
# Remover colunas irrelevantes
df = df.drop(
    columns=[
        'PassengerId',
        'Name',
        'Ticket',
        'Cabin'
    ]
)
df.head()
```

```python
# Renomear as colunas para PT-BR
df = df.rename(
    columns={
        'Survived': 'Sobreviveu',
        'Pclass': 'Classe_Passageiro',
        'Sex': 'Sexo',
        'Age': 'Idade',
        'SibSp': 'Num_irmaos_conjuges_a_bordo',
        'Parch': 'Num_pais_filhos_a_bordo',
        'Fare': 'Tarifa',
        'Embarked': 'Porto_Embarque'
    }
)
```

```python
df.head()
```

```python
# Análise de valores nulos (ausentes)
df.isnull().sum()
```

```python
# Tratamento dos valores ausentes

# Idade -> Média
df['Idade'] = df['Idade'].fillna(df['Idade'].mean())
```

```python
# Porto de Embarque -> Moda (o valor que mais aparece)
df['Porto_Embarque'] = df['Porto_Embarque'].fillna(df['Porto_Embarque'].mode()[0])
```

```python
# Converter os tipos das variáveis
df['Classe_Passageiro'] = df['Classe_Passageiro'].astype(int)
df['Sobreviveu'] = df['Sobreviveu'].astype(int)
```

```python
# Codificação das variáveis categóricas
label_encoder = LabelEncoder()

df['Sexo'] = label_encoder.fit_transform(df['Sexo'])
df['Porto_Embarque'] = label_encoder.fit_transform(df['Porto_Embarque'])

df.head()
```

# Separação entre atributos (X) e alvo (y)

--> X receber todas as colunas da nossa base
Menos a coluna "Sobreviveu"
Essas colunas são as características (features) que o modelo
utilizará para aprender os padrões

--> Y vai receber apenas a coluna "Sobreviveu"

0 = não sobreviveu

1 = sobreviveu

O algoritmo aprende com X (features) e pode prever novos casos (Y).

```python
X = df.drop('Sobreviveu', axis=1)
Y = df['Sobreviveu']

print("Atributos: ")
print(X)

print("Alvo: ")
print(Y)
```

```python
# Normalização de Escala
scaler = StandardScaler()
X_normalizado = scaler.fit_transform(X)
```

```python
X_normalizado_df = pd.DataFrame(
     X_normalizado,
     columns=X.columns
)

X_normalizado_df.head()
```

# Regressão

A regressão pode ser entendida como uma técnica estatística e de aprendizado de máquinas utilizado para prever um valor numérico contínuo.

Tenta responder:
> quanto?

> quantos?

> qual será o valor?

---

Objetivo: modelar uma relação entre uma variável dependentes (alvo, label, resposta) e uma ou mais variáveis dependentes (atributos, características).

```python
from sklearn.model_selection import train_test_split # Dividir os dados em treino e teste
from sklearn.linear_model import LogisticRegression # Regressão Logística
from sklearn.metrics import accuracy_score #Métrica de avaliação para o nosso modelo
```

# Dividir a base entre treino e teste

```python
X_train, X_teste, y_train, y_teste = train_test_split(
  X_normalizado_df, Y, test_size=0.25, random_state=42
)
```

```python
# Criar o modelo de regressão logística
modelo = LogisticRegression(max_iter = 1000)

# Treinar o nosso modelo com os dados de treino
modelo.fit(X_train, y_train)
```

```python
# Previsão no conjuto de teste
y_pred = modelo.predict(X_teste)

print("Acurácia:", accuracy_score(y_teste, y_pred))
```

```python
# Criar um novo passageiro para a base
novo_passageiro = pd.DataFrame({
    'Classe_Passageiro': [1], #Passageiro da 3ª classe
    'Sexo': [0], #1 (masc/femin)
    'Idade': [70],
    'Num_irmaos_conjuges_a_bordo': [0], #não viaja com irmão/cônjuge
    'Num_pais_filhos_a_bordo': [0], #não viaja com pais/filhos,
    'Tarifa': [15.0],
    'Porto_Embarque': [2]
})
```

```python
#Normalizar os dados do novo passageiro
novo_passageiro_normalizado = scaler.transform(novo_passageiro)
```

```python
#Predição para o novo passageiro
pred = modelo.predict(novo_passageiro_normalizado)
prob = modelo.predict_proba(novo_passageiro_normalizado)

print("\n--- Predição para o novo passageiro ---")
print("Sobreviveu? (0 = NÃO, 1 = SIM)", int(pred[0]))
print("Probabilidade de NÃO sobreviver: ", prob[0][0])
print("Probabilidade de sobreviver: ", prob[0][1])
```
